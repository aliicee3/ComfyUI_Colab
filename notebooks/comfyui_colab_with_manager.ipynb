```python
## --- ZELLE 3: START (CLOUDFLARE TUNNEL & COMFYUI) ---
## 1. Cloudflare Tunnel installieren
import os
if not os.path.exists("/usr/local/bin/cloudflared"):
  !wget https://github.com/cloudflare/cloudflared/releases/latest/download/cloudflared-linux-amd64.deb
  !dpkg -i cloudflared-linux-amd64.deb
  !rm cloudflared-linux-amd64.deb

## 2. Start-Script mit automatischer Link-Anzeige
import subprocess
import threading
import time
import socket
import re

def start_tunnel():
  while True:
    time.sleep(1)
    sock = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
    if sock.connect_ex(('127.0.0.1', 8188)) == 0:
      break
    sock.close()
    print("\n" + "="*50)
    print("✅ COMFYUI BEREIT! TUNNEL WIRD GESTARTET...")
    print("="*50 + "\n")
    !cloudflared tunnel --url http://127.0.0.1:8188

## Den Hintergrund-Prozess für den Tunnel starten
##threading.Thread(target=iframe_thread, daemon=True, args=(8188,)).start()
threading.Thread(target=start_tunnel, daemon=True).start()

## ComfyUI starten
%cd /content/ComfyUI
!python main.py
```
